```python
#importing visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot histogram for all numeric columns in the dataset
plt.figure(figsize=(12, 6))
sns.histplot(data=df, bins=20)  # Adjust the number of bins as needed
plt.title('Histogram of Entire Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# List of column names
columns = ['type', 'amount', 'nameOrig', 'oldbalanceOrg', 'newbalanceOrig', 
           'nameDest', 'oldbalanceDest', 'newbalanceDest', 'isFraud', 'isFlaggedFraud']

# Plot histograms for each column
for column in columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=column, bins=20)  # Adjust the number of bins as needed
    plt.title(f'Histogram of {column}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
label_encoder = LabelEncoder()
# Apply label encoding to the "type" column
df['type_encoded'] = label_encoder.fit_transform(df['type'])

# Display the first few rows of the DataFrame with the encoded column
print(df[['type', 'type_encoded']].head())
```
